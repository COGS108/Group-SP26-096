## Abstract

Obesity rates all across America are often blamed on fast food access, but is this assumption properly supported by data? Our project investigates whether fast-food restaurant density at the county level correlates with adult obesity rates. We utilized data from the USDA Food Environment Atlas and CDC PLACES to build a reliable dataset covering 2,947 U.S. counties, capturing fast-food density, grocery store density, median household income, and adult physical inactivity rates. Different from our analysis however, fast-food density only managed to show a weak negative correlation with obesity (r=-0.18), which only explains just 3% of county level differences in obesity rates. Overall, our findings suggest that lifestyle and socioeconomic conditions show that physical inactivity was the strongest single predictor, followed by median household income. This comes to show that fast food access alone is not the primary driver of obesity at the county level.

### USDA Food Environment Atlas

The USDA Food Environment Atlas dataset contains county level data related to food access, restaurant availability, socioeconomic conditions, and health indicators across the United States. [...]

The main variable we used from this dataset was fast-food restaurant density per 1,000 residents (FFRPTH20). This variable measured the number of fast food restaurants in relation to the cou[...]

One issue we found during cleaning the code was that some missing values were stored as placeholder values such as -9999 and -8888 instead of standard missing values. We converted these into[...]

```python
import pandas as pd
import numpy as np

# Google Drive direct-download URL (see note below on how to build it)
USDA_FILE = "https://drive.google.com/uc?export=download&id=12EUa0aqBk2-InXG2VimxGBG2g2-3Wdj0"

# Load USDA dataset
restaurants = pd.read_excel(
    USDA_FILE,
    sheet_name='RESTAURANTS'
)

restaurants.head()
```

```python
# Clean USDA restaurants dataset
restaurants.columns = restaurants.iloc[0]
restaurants = restaurants.drop(0)

restaurants_clean = restaurants[['FIPS', 'State', 'County', 'FFRPTH20']].copy()

restaurants_clean['FIPS'] = (
    restaurants_clean['FIPS']
    .astype(str)
    .str.zfill(5)
)

print(restaurants_clean.shape)

restaurants_clean.head()
```

```python
# Load & Clean Median household income

socio = pd.read_excel(
    USDA_FILE,
    sheet_name='SOCIOECONOMIC'
)
socio.columns = socio.iloc[0]
socio = socio.drop(0)

socio_clean = socio[['FIPS', 'MEDHHINC21']].copy()  # CONFIRM this column name in your file
socio_clean['FIPS'] = socio_clean['FIPS'].astype(str).str.zfill(5)

print(socio_clean.shape)

socio_clean.head()
```

```python
# Load & Clean USDA Grocery store density

stores = pd.read_excel(
    USDA_FILE,
    sheet_name='STORES'
)
stores.columns = stores.iloc[0]
stores = stores.drop(0)

stores_clean = stores[['FIPS', 'GROCPTH20']].copy()   # CONFIRM this column name in your file
stores_clean['FIPS'] = stores_clean['FIPS'].astype(str).str.zfill(5)

print(stores_clean.shape)

stores_clean.head()
```

### CDC PLACES

The CDC PLACES dataset contains county level public health data and prevalence estimates across the US. For our project, we focused mainly on adult obesity prevalence and adult physical inac[...]

The main variables we used from this dataset were (OBESITY_AdjPrev) and (LPA_AdjPrev). These estimate the percentage of adults in each county classified as obese or physically inactive. We a[...]

One limitation of this dataset is that many of the values are based on self-reported survey data, which may introduce reporting bias or inaccuracies. During the cleaning process, we found th[...]

```python
# Load CDC dataset
CDC_FILE = "https://data.cdc.gov/api/views/i46a-9kgh/rows.csv?accessType=DOWNLOAD"

cdc = pd.read_csv(
    CDC_FILE
)
```

```python
# Clean CDC dataset
cdc_clean = cdc[[
    'CountyFIPS',
    'OBESITY_AdjPrev',
    'LPA_AdjPrev'
]].copy()

cdc_clean.columns = [
    'FIPS',
    'OBESITY_AdjPrev',
    'LPA_AdjPrev'
]

cdc_clean = cdc_clean.dropna(
    subset=['OBESITY_AdjPrev', 'LPA_AdjPrev']
)

cdc_clean['FIPS'] = (
    cdc_clean['FIPS']
    .astype(str)
    .str.zfill(5)
)

print(cdc_clean.shape)
cdc_clean.head()
```

## Merge the Two Datasets into merged_df

```python
# Merge datasets using FIPS codes

# Merge all four cleaned frames on FIPS
merged_df = restaurants_clean.merge(cdc_clean, on='FIPS', how='inner')
merged_df = merged_df.merge(stores_clean, on='FIPS', how='left')   # adds GROCPTH20 - grocery store data
merged_df = merged_df.merge(socio_clean,  on='FIPS', how='left')   # adds MEDHHINC21 - median household income data

for col in ['FFRPTH20', 'GROCPTH20', 'MEDHHINC21', 'OBESITY_AdjPrev', 'LPA_AdjPrev']:
    merged_df[col] = pd.to_numeric(merged_df[col], errors='coerce').replace(
        [-9999, -8888, -9998], np.nan
    )

print(merged_df.shape)

merged_df.head()
```
