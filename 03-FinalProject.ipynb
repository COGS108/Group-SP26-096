## Background and Prior Work

Obesity is a major public health concern in the United States, affecting millions of adults and increasing the risks of chronic illnesses such as diabetes, heart disease, and hypertension. Research indicates that the prevalence of obesity varies significantly by region and is influenced by a complex set of factors.

Previous research suggests a relationship between fast-food restaurant density and obesity rates. Studies have shown that individuals living in areas with greater access to fast food are more likely to consume foods high in calories, sodium, and saturated fats, which contribute to weight gain and obesity.

The relationship between fast food availability and obesity is also affected by other socioeconomic and lifestyle factors. Lower income communities often have greater fast food availability and reduced access to healthy food options, making them particularly vulnerable to obesity. Additionally, physical inactivity is a major independent risk factor for obesity.

This project aims to examine whether fast food restaurant density is significantly associated with adult obesity rates across U.S. counties. Using county level data from the USDA Food Environment Atlas and CDC PLACES, we will investigate the predictive power of fast food density while controlling for other relevant factors including socioeconomic status and lifestyle variables.

### USDA Food Environment Atlas

The USDA Food Environment Atlas dataset contains county level data related to food access, restaurant availability, socioeconomic conditions, and health indicators across the United States. For our project, we focused on specific variables that relate to our research question.

The main variable we used from this dataset was fast-food restaurant density per 1,000 residents (FFRPTH20). This variable measured the number of fast food restaurants in relation to the county population. We also used grocery store density (GROCPTH20) and median household income (MEDHHINC21) from this dataset.

One issue we found during cleaning the code was that some missing values were stored as placeholder values such as -9999 and -8888 instead of standard missing values. We converted these into NaN values during our data cleaning process.

```python3
import pandas as pd
import numpy as np

# Google Drive direct-download URL (see note below on how to build it)
USDA_FILE = "https://drive.google.com/uc?export=download&id=12EUa0aqBk2-InXG2VimxGBG2g2-3Wdj0"

# Load USDA dataset
restaurants = pd.read_excel(
    USDA_FILE,
    sheet_name='RESTAURANTS'
)

restaurants.head()
```

```python3
# Clean USDA restaurants dataset
restaurants.columns = restaurants.iloc[0]
restaurants = restaurants.drop(0)

restaurants_clean = restaurants[['FIPS', 'State', 'County', 'FFRPTH20']].copy()

restaurants_clean['FIPS'] = (
    restaurants_clean['FIPS']
    .astype(str)
    .str.zfill(5)
)

print(restaurants_clean.shape)

restaurants_clean.head()
```

```python3
# Load & Clean Median household income

socio = pd.read_excel(
    USDA_FILE,
    sheet_name='SOCIOECONOMIC'
)
socio.columns = socio.iloc[0]
socio = socio.drop(0)

socio_clean = socio[['FIPS', 'MEDHHINC21']].copy()  # CONFIRM this column name in your file
socio_clean['FIPS'] = socio_clean['FIPS'].astype(str).str.zfill(5)

print(socio_clean.shape)

socio_clean.head()
```

```python3
# Load & Clean USDA Grocery store density

stores = pd.read_excel(
    USDA_FILE,
    sheet_name='STORES'
)
stores.columns = stores.iloc[0]
stores = stores.drop(0)

stores_clean = stores[['FIPS', 'GROCPTH20']].copy()   # CONFIRM this column name in your file
stores_clean['FIPS'] = stores_clean['FIPS'].astype(str).str.zfill(5)

print(stores_clean.shape)

stores_clean.head()
```

### CDC PLACES

The CDC PLACES dataset contains county level public health data and prevalence estimates across the US. For our project, we focused mainly on adult obesity prevalence and adult physical inactivit[...]

```python3
# Load CDC dataset
CDC_FILE = "https://data.cdc.gov/api/views/i46a-9kgh/rows.csv?accessType=DOWNLOAD"

cdc = pd.read_csv(
    CDC_FILE
)
```

```python3
# Clean CDC dataset
cdc_clean = cdc[[
    'CountyFIPS',
    'OBESITY_AdjPrev',
    'LPA_AdjPrev'
]].copy()

cdc_clean.columns = [
    'FIPS',
    'OBESITY_AdjPrev',
    'LPA_AdjPrev'
]

cdc_clean = cdc_clean.dropna(
    subset=['OBESITY_AdjPrev', 'LPA_AdjPrev']
)

cdc_clean['FIPS'] = (
    cdc_clean['FIPS']
    .astype(str)
    .str.zfill(5)
)

print(cdc_clean.shape)
cdc_clean.head()
```

## Merge the Two Datasets into merged_df

```python3
# Merge datasets using FIPS codes

# Merge all four cleaned frames on FIPS
merged_df = restaurants_clean.merge(cdc_clean, on='FIPS', how='inner')
merged_df = merged_df.merge(stores_clean, on='FIPS', how='left')   # adds GROCPTH20 - grocery store data
merged_df = merged_df.merge(socio_clean,  on='FIPS', how='left')   # adds MEDHHINC21 - median household income data

for col in ['FFRPTH20', 'GROCPTH20', 'MEDHHINC21', 'OBESITY_AdjPrev', 'LPA_AdjPrev']:
    merged_df[col] = pd.to_numeric(merged_df[col], errors='coerce').replace(
        [-9999, -8888, -9998], np.nan
    )

print(merged_df.shape)

merged_df.head()
```
