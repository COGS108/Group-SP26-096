## Background and Prior Work

Obesity is a major public health concern in the United States, affecting millions of adults and increasing the risks of chronic illnesses such as diabetes, heart disease, and hypertension. Researchers have increasingly studied how environmental factors influence dietary behavior and long-term health outcomes. One important concept is the food environment, and refers to the availability and accessibility of different food options within a geographic area. Areas with a high concentration of fast-food restaurants are often associated with increased consumption of highly caloric but nutrient poor foods.

Previous research suggests a relationship between fast-food restaurant density and obesity rates. Studies have shown that individuals living in areas with greater access to fast food are more likely to consume high-calorie meals and have higher body mass index (BMI) values. Fraser et al. (2010) found that proximity to fast food outlets is positively associated with increased fast food consumption and obesity risk. The Centers for Disease Control and Prevention (CDC) also reports that environmental and community factors, including access to unhealthy food options, playing a significant role in obesity trends.

The relationship between fast food availability and obesity is also affected by other socioeconomic and lifestyle factors. Lower income communities often have greater fast food availability and reduced access to grocery stores that provide healthier food options. Larson et al. (2009) found that limited supermarket access is associated with poorer diet quality and higher obesity rates. Physical inactivity may also contribute to obesity trends, especially in areas with fewer recreational opportunities or reduced walkability.

This project aims to examine whether fast food restaurant density is significantly associated with adult obesity rates across U.S. counties. Using county level data from the USDA Food Environment Atlas and CDC PLACES datasets, we will also analyze how variables such as grocery store access, median household income, and physical inactivity may influence this relationship.

### USDA Food Environment Atlas

The USDA Food Environment Atlas dataset contains county level data related to food access, restaurant availability, socioeconomic conditions, and health indicators across the United States. For our project, we focused mainly on the "RESTAURANTS" sheet, as it includes variables measuring fast food restaurant density, which is one of the main variables connected to our research question.

The main variable we used from this dataset was fast-food restaurant density per 1,000 residents (FFRPTH20). This variable measured the number of fast food restaurants in relation to the county population size in 2020. We also used county FIPS codes because they allow the dataset to be merged with other county level datasets.

One issue we found during cleaning the code was that some missing values were stored as placeholder values such as -9999 and -8888 instead of standard missing values. We converted these into NaN values during the wrangling process so they would not affect analysis or summary statistics. Another limitation is that the data is collected only at the county level so we cannot look fully at the differences within individual counties or explain counties' individual behavior.

```python
import pandas as pd
import numpy as np

# Google Drive direct-download URL (see note below on how to build it)
USDA_FILE = "https://drive.google.com/uc?export=download&id=12EUa0aqBk2-InXG2VimxGBG2g2-3Wdj0"

# Load USDA dataset
restaurants = pd.read_excel(
    USDA_FILE,
    sheet_name='RESTAURANTS'
)

restaurants.head()
```

```python
# Clean USDA restaurants dataset
restaurants.columns = restaurants.iloc[0]
restaurants = restaurants.drop(0)

restaurants_clean = restaurants[['FIPS', 'State', 'County', 'FFRPTH20']].copy()

restaurants_clean['FIPS'] = (
    restaurants_clean['FIPS']
    .astype(str)
    .str.zfill(5)
)

print(restaurants_clean.shape)

restaurants_clean.head()
```

```python
# Load & Clean Median household income

socio = pd.read_excel(
    USDA_FILE,
    sheet_name='SOCIOECONOMIC'
)
socio.columns = socio.iloc[0]
socio = socio.drop(0)

socio_clean = socio[['FIPS', 'MEDHHINC21']].copy()  # CONFIRM this column name in your file
socio_clean['FIPS'] = socio_clean['FIPS'].astype(str).str.zfill(5)

print(socio_clean.shape)

socio_clean.head()
```

```python
# Load & Clean USDA Grocery store density

stores = pd.read_excel(
    USDA_FILE,
    sheet_name='STORES'
)
stores.columns = stores.iloc[0]
stores = stores.drop(0)

stores_clean = stores[['FIPS', 'GROCPTH20']].copy()   # CONFIRM this column name in your file
stores_clean['FIPS'] = stores_clean['FIPS'].astype(str).str.zfill(5)

print(stores_clean.shape)

stores_clean.head()
```

### CDC PLACES

The CDC PLACES dataset contains county level public health data and prevalence estimates across the US. For our project, we focused mainly on adult obesity prevalence and adult physical inactivity prevalence because both variables are directly related to our research question and hypothesis.

The main variables we used from this dataset were (OBESITY_AdjPrev) and (LPA_AdjPrev). These estimate the percentage of adults in each county classified as obese or physically inactive. We also used counties FIPS code for this data set as well so we could merge it with the USDA Food Environment Atlas dataset.

One limitation of this dataset is that many of the values are based on self-reported survey data, which may introduce reporting bias or inaccuracies. During the cleaning process, we found that some counties had missing obesity or physical inactivity estimates, so those rows were removed from the cleaned dataframe before analysis.

```python
# Load CDC dataset
CDC_FILE = "https://data.cdc.gov/api/views/i46a-9kgh/rows.csv?accessType=DOWNLOAD"

cdc = pd.read_csv(
    CDC_FILE
)
```

```python
# Clean CDC dataset

cdc_clean = cdc[[
    'CountyFIPS',
    'StateAbbr',
    'CountyName',
    'OBESITY_AdjPrev',
    'LPA_AdjPrev'
]].copy()

cdc_clean.columns = [
    'FIPS',
    'State',
    'County',
    'obesity_rate',
    'physical_inactivity_rate'
]

cdc_clean = cdc_clean.dropna(
    subset=['obesity_rate', 'physical_inactivity_rate']
)

cdc_clean['FIPS'] = (
    cdc_clean['FIPS']
    .astype(str)
    .str.zfill(5)
)

print(cdc_clean.shape)

cdc_clean.head()
```
