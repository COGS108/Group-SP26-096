## Background and Prior Work

Obesity is a major public health concern in the United States, affecting millions of adults and increasing the risks of chronic illnesses such as diabetes, heart disease, and hypertension. Researchers have increasingly studied how environmental factors influence dietary behavior and long-term health outcomes. One important concept is the food environment, and refers to the availability and accessibility of different food options within a geographic area. Areas with a high concentration of fast-food restaurants are often associated with increased consumption of highly caloric but nutrient poor foods.

Previous research suggests a relationship between fast-food restaurant density and obesity rates. Studies have shown that individuals living in areas with greater access to fast food are more likely to consume high-calorie meals and have higher body mass index (BMI) values. Fraser et al. (2010) found that proximity to fast food outlets is positively associated with increased fast food consumption and obesity risk. The Centers for Disease Control and Prevention (CDC) also reports that environmental and community factors, including access to unhealthy food options, playing a significant role in obesity trends.

The relationship between fast food availability and obesity is also affected by other socioeconomic and lifestyle factors. Lower income communities often have greater fast food availability and reduced access to grocery stores that provide healthier food options. Larson et al. (2009) found that limited supermarket access is associated with poorer diet quality and higher obesity rates. Physical inactivity may also contribute to obesity trends, especially in areas with fewer recreational opportunities or reduced walkability.

This project aims to examine whether fast food restaurant density is significantly associated with adult obesity rates across U.S. counties. Using county level data from the USDA Food Environment Atlas and CDC PLACES datasets, we will also analyze how variables such as grocery store access, median household income, and physical inactivity may influence this relationship.

### USDA Food Environment Atlas

The USDA Food Environment Atlas dataset contains county level data related to food access, restaurant availability, socioeconomic conditions, and health indicators across the United States. For our project, we focused mainly on the "RESTAURANTS" sheet, as it includes variables measuring fast food restaurant density, which is one of the main variables connected to our research question.

The main variable we used from this dataset was fast-food restaurant density per 1,000 residents (FFRPTH20). This variable measured the number of fast food restaurants in relation to the county population size in 2020. We also used county FIPS codes because they allow the dataset to be merged with other county level datasets.

One issue we found during cleaning the code was that some missing values were stored as placeholder values such as -9999 and -8888 instead of standard missing values. We converted these into NaN values during the wrangling process so they would not affect analysis or summary statistics. Another limitation is that the data is collected only at the county level so we cannot look fully at the differences within individual counties or explain counties' individual behavior.

```python
import pandas as pd
import numpy as np

# Google Drive direct-download URL (see note below on how to build it)
USDA_FILE = "https://drive.google.com/uc?export=download&id=12EUa0aqBk2-InXG2VimxGBG2g2-3Wdj0"

# Load USDA dataset
restaurants = pd.read_excel(
    USDA_FILE,
    sheet_name='RESTAURANTS'
)

restaurants.head()
```

```python
# Clean USDA restaurants dataset
restaurants.columns = restaurants.iloc[0]
restaurants = restaurants.drop(0)

restaurants_clean = restaurants[['FIPS', 'State', 'County', 'FFRPTH20']].copy()

restaurants_clean['FIPS'] = (
    restaurants_clean['FIPS']
    .astype(str)
    .str.zfill(5)
)

print(restaurants_clean.shape)

restaurants_clean.head()
```

```python
# Load & Clean Median household income

socio = pd.read_excel(
    USDA_FILE,
    sheet_name='SOCIOECONOMIC'
)
socio.columns = socio.iloc[0]
socio = socio.drop(0)

socio_clean = socio[['FIPS', 'MEDHHINC21']].copy()  # CONFIRM this column name in your file
socio_clean['FIPS'] = socio_clean['FIPS'].astype(str).str.zfill(5)

print(socio_clean.shape)

socio_clean.head()
```

```python
# Load & Clean USDA Grocery store density

stores = pd.read_excel(
    USDA_FILE,
    sheet_name='STORES'
)
stores.columns = stores.iloc[0]
stores = stores.drop(0)

stores_clean = stores[['FIPS', 'GROCPTH20']].copy()   # CONFIRM this column name in your file
stores_clean['FIPS'] = stores_clean['FIPS'].astype(str).str.zfill(5)

print(stores_clean.shape)

stores_clean.head()
```

### CDC PLACES

The CDC PLACES dataset contains county level public health data and prevalence estimates across the US. For our project, we focused mainly on adult obesity prevalence and adult physical inactivity prevalence because both variables are directly related to our research question and hypothesis.

The main variables we used from this dataset were (OBESITY_AdjPrev) and (LPA_AdjPrev). These estimate the percentage of adults in each county classified as obese or physically inactive. We also used counties FIPS code for this data set as well so we could merge it with the USDA Food Environment Atlas dataset.

One limitation of this dataset is that many of the values are based on self-reported survey data, which may introduce reporting bias or inaccuracies. During the cleaning process, we found that some counties had missing obesity or physical inactivity estimates, so those rows were removed from the cleaned dataframe before analysis.

```python
# Load CDC dataset
CDC_FILE = "https://data.cdc.gov/api/views/i46a-9kgh/rows.csv?accessType=DOWNLOAD"

cdc = pd.read_csv(
    CDC_FILE
)
```

```python
# Clean CDC dataset
cdc_clean = cdc[[
    'CountyFIPS',
    'OBESITY_AdjPrev',
    'LPA_AdjPrev'
]].copy()

cdc_clean.columns = [
    'FIPS',
    'OBESITY_AdjPrev',
    'LPA_AdjPrev'
]

cdc_clean = cdc_clean.dropna(
    subset=['OBESITY_AdjPrev', 'LPA_AdjPrev']
)

cdc_clean['FIPS'] = (
    cdc_clean['FIPS']
    .astype(str)
    .str.zfill(5)
)

print(cdc_clean.shape)
cdc_clean.head()
```

## Merge the Two Datasets into merged_df

```python
# Merge datasets using FIPS codes

# Merge all four cleaned frames on FIPS
merged_df = restaurants_clean.merge(cdc_clean, on='FIPS', how='inner')
merged_df = merged_df.merge(stores_clean, on='FIPS', how='left')   # adds GROCPTH20 - grocery store data
merged_df = merged_df.merge(socio_clean,  on='FIPS', how='left')   # adds MEDHHINC21 - median household income data

for col in ['FFRPTH20', 'GROCPTH20', 'MEDHHINC21', 'OBESITY_AdjPrev', 'LPA_AdjPrev']:
    merged_df[col] = pd.to_numeric(merged_df[col], errors='coerce').replace(
        [-9999, -8888, -9998], np.nan
    )

print(merged_df.shape)

merged_df.head()
```

### Exploratory Data Analysis

In this section, we perform exploratory data analysis on our merged county-level dataset to examine variable distributions, bivariate relationships, and the influence of additional factors on adult obesity rates. We load the cleaned data directly from merged_df: the two datasets cleaned and merged.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', font_scale=1.05)
BLUE   = '#2C7BB6'
ORANGE = '#D7191C'
GREEN  = '#1A9641'
GRAY   = '#636363'
PURPLE = '#7B3294'   # added a 5th color

df = merged_df.copy() # grab df from earlier instead of importing again
print(f"Counties in analysis: {len(df):,}")

key_vars = {
    'FFRPTH20':        'Fast Food Density\n(per 1k residents)',
    'OBESITY_AdjPrev': 'Adult Obesity\nPrevalence (%)',
    'LPA_AdjPrev':     'Physical Inactivity\nPrevalence (%)',
    'MEDHHINC21':      'Median Household\nIncome ($)',     # name change MEDHHINC15 -> MEDHHINC21
    'GROCPTH20':       'Grocery Store\nDensity (per 1k)',  # name change GROC14 -> GROCPTH20
}
available = [c for c in key_vars if c in df.columns]
display(df[available].agg(['mean','median','std','skew']).round(2))
```

```python
fig, axes = plt.subplots(1, len(available), figsize=(5 * len(available), 4.5))
colors = [BLUE, ORANGE, GREEN, GRAY, PURPLE]   # added 5th color

for ax, col, color in zip(axes, available, colors):
    data = df[col].dropna()
    ax.hist(data, bins=45, color=color, edgecolor='white', alpha=0.85, linewidth=0.4)
    ax.axvline(data.mean(),   color='black',  linestyle='--', linewidth=1.4,
               label=f'Mean: {data.mean():.2f}')
    ax.axvline(data.median(), color='tomato', linestyle='-',  linewidth=1.8,
               label=f'Median: {data.median():.2f}')
    ax.set_title(key_vars[col], fontsize=10.5, pad=8)
    ax.set_xlabel('Value', fontsize=9)
    ax.set_ylabel('Number of Counties', fontsize=9)
    ax.legend(fontsize=8.5)
    ax.tick_params(labelsize=8)

plt.suptitle('Distribution of Key Variables Across U.S. Counties',
             fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('data/02-processed/eda_fig1_distributions.png', dpi=150, bbox_inches='tight')
plt.show()
```

Fast food restaurant density is notably right-skewed: most counties have low density, but a long right tail of high-density outlier counties pulls the mean above the median. Adult obesity and physical inactivity are closer to normally distributed. Because of the skew in fast food density, we focus on the median as the better measure of central tendency for that variable and will verify that our correlation results are not driven by outlier counties.

#### Section 2: Fast Food Density vs. Obesity — Primary Research Question

Our primary research question asks whether counties with higher fast food restaurant density also have higher adult obesity rates. We examine this directly through a scatterplot with a trend line, and compare it side-by-side with the physical inactivity–obesity relationship to begin assessing which variable is the stronger predictor (research question 3).

```python
pairs = [
    ('FFRPTH20',    'Fast Food Density (per 1k residents)',  BLUE),
    ('LPA_AdjPrev', 'Physical Inactivity Prevalence (%)',    ORANGE),
]
pairs = [(x, lbl, c) for x, lbl, c in pairs if x in df.columns]

fig, axes = plt.subplots(1, len(pairs), figsize=(6.5 * len(pairs), 5))
if len(pairs) == 1:
    axes = [axes]

for ax, (xcol, xlabel, color) in zip(axes, pairs):
    subset = df[[xcol, 'OBESITY_AdjPrev']].dropna()
    x, y   = subset[xcol], subset['OBESITY_AdjPrev']

    ax.scatter(x, y, alpha=0.25, s=10, color=color, rasterized=True)

    m, b   = np.polyfit(x, y, 1)
    x_line = np.linspace(x.min(), x.max(), 200)
    ax.plot(x_line, m * x_line + b, color='black', linewidth=1.8,
            label=f'Trend (slope = {m:.2f})')

    r = np.corrcoef(x, y)[0, 1]
    ax.text(0.97, 0.05, f'r = {r:.3f}', transform=ax.transAxes,
            ha='right', va='bottom', fontsize=10.5,
            bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8))

    ax.set_xlabel(xlabel, fontsize=10)
    ax.set_ylabel('Adult Obesity Prevalence (%)', fontsize=10)
    ax.legend(fontsize=9)
    ax.tick_params(labelsize=8.5)

axes[0].set_title('Fast Food Density vs. Adult Obesity', fontsize=11, pad=6)
if len(axes) > 1:
    axes[1].set_title('Physical Inactivity vs. Adult Obesity', fontsize=11, pad=6)

plt.suptitle('Bivariate Relationships with Adult Obesity Prevalence (U.S. Counties)',
             fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('data/02-processed/eda_fig2_scatterplots.png', dpi=150, bbox_inches='tight')
plt.show()
```

The two variables (Fast food density and Physical Inactivity) are pointing in opposite directions. Physical inactivity is strongly, positively correlated with adult obesity [r = 0.75]: there is a clear upward trend seen in the graph. Fast food density has a weaker negative association with adult obesity [r = -0.18]. This suggests that more fast food restaurant densities tend to have lower obesity levels, not higher obesity levels like we hypothesized. 

This preliminary comparison shows that physical inactivity is a stronger predictor of obesity than fast food density alone, which we will confirm with regression analysis. This challenges our hypothesis that higher density fast food restaurant counties will tend to have higher adult obesity rates. But this aligns with our hypothesis that fast food density alone would not be able to accurately predict adult obesity rates by US county. 

#### Section 3: Influence of Additional Factors — Income, Grocery Access, and All-Variable Correlations

Our second research question asks how additional factors like median household income and grocery store access influence the fast food–obesity relationship. Here we use a correlation heatmap to examine all pairwise relationships simultaneously, revealing which variables are most strongly associated with obesity and whether any predictors are highly intercorrelated.

```python
heatmap_vars = [c for c in ['FFRPTH20', 'OBESITY_AdjPrev', 'LPA_AdjPrev',
                            'MEDHHINC21', 'GROCPTH20'] if c in df.columns]   # Name changes: MEDHHINC15->MEDHHINC21, GROC14->GROCPTH20

rename_map = {
    'FFRPTH20':        'Fast Food\nDensity',
    'OBESITY_AdjPrev': 'Adult\nObesity (%)',
    'LPA_AdjPrev':     'Physical\nInactivity (%)',
    'MEDHHINC21':      'Median\nIncome ($)',        # was MEDHHINC15
    'GROCPTH20':       'Grocery Store\nDensity',     # was GROC14
}

corr_matrix  = df[heatmap_vars].corr()
corr_display = corr_matrix.rename(index=rename_map, columns=rename_map)
mask = np.eye(len(corr_matrix), dtype=bool)

fig, ax = plt.subplots(figsize=(7, 5.5))
sns.heatmap(
    corr_display,
    annot=True, fmt='.2f',
    cmap='RdBu_r', center=0, vmin=-1, vmax=1,
    mask=mask,
    linewidths=0.6, linecolor='white',
    annot_kws={'size': 11, 'weight': 'bold'},
    cbar_kws={'label': 'Pearson r', 'shrink': 0.8},
    ax=ax
)
ax.set_title('Correlation Matrix of County-Level Variables',
             fontsize=12, fontweight='bold', pad=10)
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha='right', fontsize=9)
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, fontsize=9)
plt.tight_layout()
plt.savefig('data/02-processed/eda_fig3_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```

### Regression Analysis

The correlation heatmaps visualizes the pairwise relationships between each variables, but it doesn't answer the question we asked: can fast food density predict obesity **while accounting for other factors**? 

To answer this question, we fit two least square regression models.
1. Model A fits adult obesity from fast food density alone. After seeing the previous correlation data, we expect this to be a poor fit.
2. Model B fits adult obesity from fast food density, but adds **physical inactivity, median household income, and grocery store density as controls**.

We can evaluate the originally asked question by comparing the two models' coefficients and seeing if it changes after adding the control variables. 

```python
import statsmodels.formula.api as smf

# dropna missing cells
model_df = df[['OBESITY_AdjPrev', 'FFRPTH20', 'LPA_AdjPrev',
               'MEDHHINC21', 'GROCPTH20']].dropna()
print(f"Counties used in regression: {len(model_df)}")

# Model A: BASELINE: fast food density ONLY
model_a = smf.ols('OBESITY_AdjPrev ~ FFRPTH20', data=model_df).fit()
print(model_a.summary())
```

## Model A Regression (ONLY Fast Food Density)
As expected, the Model A regression shows that fast food density as a standalone variables is a very poor predictor of adult obesity rates [coefficient is **negative** and statistically significant: p < 0.0001, but the R-squared value is **only** 0.029]. The R squared value estimates that fast food density only explains around 3% of the variation in US county obesity rates. We also see a negative coefficient when we hypothesized a positive correlation between the two variables: we expected to see that higher fast food density counties tend to have higher adult obesity rates. The tiny R squared value itself tells us that fast food density should not be considered a reliable predictor for adult obesity in US counties. 

```python
# Model B: control variables accounted for
model_b = smf.ols(
    'OBESITY_AdjPrev ~ FFRPTH20 + LPA_AdjPrev + MEDHHINC21 + GROCPTH20',
    data=model_df
).fit()
print(model_b.summary())
```

## Model B Regression (3 control variables added)
We see that after considering the 3 new variables: **physical inactivity, median household income, and grocery store density**, the R-squared value of the regression grows from 0.029 to 0.576. The combination of variables is estimated to explain 57.6% of the variability in adult obesity rates in US counties. Model B, with the control variables, is a far better predictor for adult obesity rates compared to Model A. 

Physical inactivity has a strong, significant positive correlation with adult obesity (+0.52, p < 0.0001). This was expected. Median household income has a small, significant negative correlation (-7.27e-5, p < 0.0001), which was also expected. 

```python
# Comparing the two coefficients
print("\nFast food coefficient alone:        ", round(model_a.params['FFRPTH20'], 3))
print("Fast food coefficient with controls:", round(model_b.params['FFRPTH20'], 3))
```

The fast food correlation coefficient shrunk from -3.47 to -1.26 after the introduction of the control variables. The fast food density variable was never positive and became a weaker predictor once other variables were considered. This data does not support our hypothesis that higher fast food restaurant densities would be a predictor for adult obesity rates. 

Moreover, we hypothesize that this improvement in regression fit was due to the introduction of physical inactivity and median household income. We saw the strong correlation between those three variables earlier, and they continue to be strongly correlated. 

## Accurately Comparing Predictor Strength

Its important to note that the coefficients of the variables in these regressions are not directly comparable. The fast food data is labelled using restaurants per 1000 people while inactivity rate is a percentage. We want to accurately find the strongest predictor, so the following cell will standardize each variable using a z-score metric. This refitting of the models will allow us to compare the standardized coefficients. 

To wrap up the fast food density and obesity correlation analyses, we will also report the Spearman rank correlation. This ensures that the results that we saw were not due to outlier counties (as seen by the right-skew seen in the distribution). 

```python
from scipy import stats

# Standardize all variables on z-scores
z = (model_df - model_df.mean()) / model_df.std()
model_z = smf.ols(
    'OBESITY_AdjPrev ~ FFRPTH20 + LPA_AdjPrev + MEDHHINC21 + GROCPTH20',
    data=z
).fit()

# Standardized coefficients, ranked by absolute size
std_coefs = model_z.params.drop('Intercept')
print("Standardized coefficients (larger magnitude = stronger predictor):")
print(std_coefs.reindex(std_coefs.abs().sort_values(ascending=False).index).round(3))

# Skew-robust check on the primary relationship
pearson  = model_df['FFRPTH20'].corr(model_df['OBESITY_AdjPrev'])
spearman = stats.spearmanr(model_df['FFRPTH20'], model_df['OBESITY_AdjPrev']).correlation
print(f"\nFast food vs obesity:  Pearson r = {pearson:.3f} | Spearman rho = {spearman:.3f}")
```

As expected, physical inactivity stands as the strongest predictor of adult obesity (+0.557 standardized coefficient). It continues showing a strong positive relationship. Median household income comes after (-0.239), which continue showing a moderate negative relationship. Grocery store density (-0.071) and Fast Food Density (-0.062) continue to be very weak, negative predictors. 

This shows that **physical inactivity**, not fast food restaurant density, is the strongest predictor of adult obesity rate in our dataset. 

The Spearman rank correlation between fast food density and obesity: rho = -0.141 is very close to the computed pearson r value: r = -0.171. The agreement between these two values confirm that the slight negative relationship is actually real in the data and not a byproduct of outliers. The Spearman rank test is rank-based, so it is able to compute correlation despite outliers.
