## Research Question

Does living in a high fast-food restaurant density area correlate with adult obesity rates in the United States? 
- How do factors like household income, physical activity, and type of living community affect this relationship?
- How does access to grocery stores alter this relationship?
- Which is the strongest predictor of obesity in this dataset?



## Background and Prior Work

Obesity is a major public health concern in the United States, affecting over 40% of adults and increasing the risk of chronic diseases such as diabetes, heart disease, and hypertension. Researchers have increasingly focused on how environmental factors, particularly access to food, influence dietary behavior and long-term health outcomes. One key concept is the “food environment,” which refers to the availability and accessibility of different types of food options within a geographic area. Areas with a high concentration of fast-food restaurants are often associated with increased consumption of calorie-dense, nutrient-poor foods.

Previous research suggests a relationship between fast-food restaurant density and obesity rates. Studies have shown that individuals living in areas with greater access to fast food are more likely to consume high-calorie meals and have higher body mass index (BMI) values. For example, research has found that proximity to fast-food outlets is positively associated with increased fast-food consumption and obesity risk (Fraser et al., 2010). The Centers for Disease Control and Prevention (CDC) also reports that environmental and community factors, including access to unhealthy food options, play a significant role in shaping obesity trends.

However, the relationship between fast-food availability and obesity is not purely direct. Socioeconomic factors such as income, education, and neighborhood characteristics often influence both food access and health outcomes. Lower-income communities tend to have higher densities of fast-food restaurants and fewer grocery stores offering fresh, healthy foods. Larson et al. (2009) found that limited access to supermarkets is associated with poorer diet quality and higher obesity rates, suggesting that food environment disparities contribute to health inequalities.

In addition to food access, lifestyle factors such as physical activity levels also play a critical role in obesity. Areas with fewer recreational spaces or lower walkability may contribute to sedentary behavior, compounding the effects of poor diet. Research indicates that both dietary environment and physical activity must be considered together when analyzing obesity trends, as they interact to influence overall health outcomes.

While prior studies establish connections between food environment and obesity, there is still debate about the strength of these relationships when multiple variables are considered simultaneously. Some studies suggest that fast-food density alone is not a strong predictor once factors like income and lifestyle are controlled for. Therefore, this project aims to examine whether fast-food restaurant density is significantly correlated with adult obesity rates in the United States, and how additional variables such as income, physical activity, and access to grocery stores influence this relationship.

## Hypothesis


We predict a strong correlation between higher fast food restaurant density areas and higher adult obesity rates. We also hypothesize that inclusion of other variables like average household income and physical activity rates in these areas would moderate this relationship. 

This prediction is based on prior datasets and research done on comparing access to food options and health statistics. Previous work has shown that greater exposure to fast food options directly affects the likelihood that residents living in those areas will consume calorie-dense, low-nutrition foods. We also see that fast food restaurants tend to cluster disproportionately in lower income areas. Building on previous knowledge, we would like to confirm that 'food environment' has a genuine correlation with resident health outcomes, specifically obesity.
