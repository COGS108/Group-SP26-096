## Background and Prior Work

Obesity is a major public health concern in the United States, affecting millions of adults and increasing the risks of chronic illnesses such as diabetes, heart disease, and hypertension. Researchers have increasingly studied how environmental factors influence dietary behavior and long-term health outcomes. One important concept is the food environment, and refers to the availability and accessibility of different food options within a geographic area. Areas with a high concentration of fast-food restaurants are often associated with increased consumption of highly caloric but nutrient poor foods.

Previous research suggests a relationship between fast-food restaurant density and obesity rates. Studies have shown that individuals living in areas with greater access to fast food are more likely to consume high-calorie meals and have higher body mass index (BMI) values. Fraser et al. (2010) found that proximity to fast food outlets is positively associated with increased fast food consumption and obesity risk. The Centers for Disease Control and Prevention (CDC) also reports that environmental and community factors, including access to unhealthy food options, playing a significant role in obesity trends.

The relationship between fast food availability and obesity is also affected by other socioeconomic and lifestyle factors. Lower income communities often have greater fast food availability and reduced access to grocery stores that provide healthier food options. Larson et al. (2009) found that limited supermarket access is associated with poorer diet quality and higher obesity rates. Physical inactivity may also contribute to obesity trends, especially in areas with fewer recreational opportunities or reduced walkability.

This project aims to examine whether fast food restaurant density is significantly associated with adult obesity rates across U.S. counties. Using county level data from the USDA Food Environment Atlas and CDC PLACES datasets, we will also analyze how variables such as grocery store access, median household income, and physical inactivity may influence this relationship.

## Hypothesis

We predict that counties with higher fast food restaurants will also have higher adult obesity rates. We also hypothesize that additional variables such as median household income, grocery store access, and adult physical inactivity will influence the strength of this relationship.

This prediction is based on previous research suggesting that greater access to fast food increases the likelihood of consuming high calorie, low nutrition foods. Prior studies have also shown that lower income communities often have greater fast food availability and fewer healthy food options. Because obesity is influenced by multiple environmental and socioeconomic factors, we expect that fast food density alone will not fully explain obesity rates, but will remain an important contributing factor.

## Results

### Exploratory Data Analysis

In this section, we perform exploratory data analysis on our merged county-level dataset to examine variable distributions, bivariate relationships, and the influence of additional factors on adult obesity rates. We load the cleaned data directly from `data/02-processed` so this section runs independently of the wrangling steps above.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', font_scale=1.05)
BLUE   = '#2C7BB6'
ORANGE = '#D7191C'
GREEN  = '#1A9641'
GRAY   = '#636363'

df = pd.read_csv('data/02-processed/merged_county_data.csv')
print(f"Counties in analysis: {len(df):,}")

key_vars = {
    'FFRPTH20':        'Fast Food Density\n(per 1k residents)',
    'OBESITY_AdjPrev': 'Adult Obesity\nPrevalence (%)',
    'LPA_AdjPrev':     'Physical Inactivity\nPrevalence (%)',
    # 'MEDHHINC15':    'Median Household\nIncome ($)',
    # 'GROC14':        'Grocery Store\nDensity (per 1k)',
}
available = [c for c in key_vars if c in df.columns]
display(df[available].agg(['mean','median','std','skew']).round(2))
```

#### Section 1: Distributions of Key Variables

Before examining relationships between variables, we inspect the distribution of each key variable to understand shape, spread, and skewness. This is important because skewed distributions affect our choice of summary statistics and inform how we interpret correlations. Both mean and median are annotated so the impact of skewness is immediately visible.

```python
fig, axes = plt.subplots(1, len(available), figsize=(5 * len(available), 4.5))
colors = [BLUE, ORANGE, GREEN, GRAY]

for ax, col, color in zip(axes, available, colors):
    data = df[col].dropna()
    ax.hist(data, bins=45, color=color, edgecolor='white', alpha=0.85, linewidth=0.4)
    ax.axvline(data.mean(),   color='black',  linestyle='--', linewidth=1.4,
               label=f'Mean: {data.mean():.2f}')
    ax.axvline(data.median(), color='tomato', linestyle='-',  linewidth=1.8,
               label=f'Median: {data.median():.2f}')
    ax.set_title(key_vars[col], fontsize=10.5, pad=8)
    ax.set_xlabel('Value', fontsize=9)
    ax.set_ylabel('Number of Counties', fontsize=9)
    ax.legend(fontsize=8.5)
    ax.tick_params(labelsize=8)

plt.suptitle('Distribution of Key Variables Across U.S. Counties',
             fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('data/02-processed/eda_fig1_distributions.png', dpi=150, bbox_inches='tight')
plt.show()
```

Fast food restaurant density is notably right-skewed — most counties have low density, but a long right tail of high-density outlier counties pulls the mean above the median. Adult obesity and physical inactivity are closer to normally distributed. Because of the skew in fast food density, we focus on the median as the better measure of central tendency for that variable and will verify that our correlation results are not driven by outlier counties.

#### Section 2: Fast Food Density vs. Obesity — Primary Research Question

Our primary research question asks whether counties with higher fast food restaurant density also have higher adult obesity rates. We examine this directly through a scatterplot with a trend line, and compare it side-by-side with the physical inactivity–obesity relationship to begin assessing which variable is the stronger predictor (research question 3).

```python
pairs = [
    ('FFRPTH20',    'Fast Food Density (per 1k residents)',  BLUE),
    ('LPA_AdjPrev', 'Physical Inactivity Prevalence (%)',    ORANGE),
]
pairs = [(x, lbl, c) for x, lbl, c in pairs if x in df.columns]

fig, axes = plt.subplots(1, len(pairs), figsize=(6.5 * len(pairs), 5))
if len(pairs) == 1:
    axes = [axes]

for ax, (xcol, xlabel, color) in zip(axes, pairs):
    subset = df[[xcol, 'OBESITY_AdjPrev']].dropna()
    x, y   = subset[xcol], subset['OBESITY_AdjPrev']

    ax.scatter(x, y, alpha=0.25, s=10, color=color, rasterized=True)

    m, b   = np.polyfit(x, y, 1)
    x_line = np.linspace(x.min(), x.max(), 200)
    ax.plot(x_line, m * x_line + b, color='black', linewidth=1.8,
            label=f'Trend (slope = {m:.2f})')

    r = np.corrcoef(x, y)[0, 1]
    ax.text(0.97, 0.05, f'r = {r:.3f}', transform=ax.transAxes,
            ha='right', va='bottom', fontsize=10.5,
            bbox=dict(boxstyle='round,pad=0.3', facecolor='white', alpha=0.8))

    ax.set_xlabel(xlabel, fontsize=10)
    ax.set_ylabel('Adult Obesity Prevalence (%)', fontsize=10)
    ax.legend(fontsize=9)
    ax.tick_params(labelsize=8.5)

axes[0].set_title('Fast Food Density vs. Adult Obesity', fontsize=11, pad=6)
if len(axes) > 1:
    axes[1].set_title('Physical Inactivity vs. Adult Obesity', fontsize=11, pad=6)

plt.suptitle('Bivariate Relationships with Adult Obesity Prevalence (U.S. Counties)',
             fontsize=13, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('data/02-processed/eda_fig2_scatterplots.png', dpi=150, bbox_inches='tight')
plt.show()
```

Both predictors show a positive association with adult obesity, but physical inactivity has a noticeably tighter, steeper relationship such as its points cluster closer to the trend line and its Pearson r is higher. Fast food density shows a weaker, more diffuse association. This preliminary comparison suggests physical inactivity may be a stronger predictor of obesity than fast food density alone, which we will confirm with formal analysis. It also supports our hypothesis that fast food density does not fully explain obesity rates on its own.

#### Section 3: Influence of Additional Factors — Income, Grocery Access, and All-Variable Correlations

Our second research question asks how additional factors like median household income and grocery store access influence the fast food–obesity relationship. Here we use a correlation heatmap to examine all pairwise relationships simultaneously, revealing which variables are most strongly associated with obesity and whether any predictors are highly intercorrelated.

```python
# Update column names to match your actual processed dataset
# e.g. 'MEDHHINC15' for income, 'GROC14' for grocery store density
heatmap_vars = [c for c in ['FFRPTH20', 'OBESITY_AdjPrev', 'LPA_AdjPrev',
                             'MEDHHINC15', 'GROC14'] if c in df.columns]

rename_map = {
    'FFRPTH20':        'Fast Food\nDensity',
    'OBESITY_AdjPrev': 'Adult\nObesity (%)',
    'LPA_AdjPrev':     'Physical\nInactivity (%)',
    'MEDHHINC15':      'Median\nIncome ($)',
    'GROC14':          'Grocery Store\nDensity',
}

corr_matrix  = df[heatmap_vars].corr()
corr_display = corr_matrix.rename(index=rename_map, columns=rename_map)
mask = np.eye(len(corr_matrix), dtype=bool)

fig, ax = plt.subplots(figsize=(7, 5.5))
sns.heatmap(
    corr_display,
    annot=True, fmt='.2f',
    cmap='RdBu_r', center=0, vmin=-1, vmax=1,
    mask=mask,
    linewidths=0.6, linecolor='white',
    annot_kws={'size': 11, 'weight': 'bold'},
    cbar_kws={'label': 'Pearson r', 'shrink': 0.8},
    ax=ax
)
ax.set_title('Correlation Matrix of County-Level Variables',
             fontsize=12, fontweight='bold', pad=10)
ax.set_xticklabels(ax.get_xticklabels(), rotation=30, ha='right', fontsize=9)
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, fontsize=9)
plt.tight_layout()
plt.savefig('data/02-processed/eda_fig3_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
```

The heatmap provides a full picture of how every variable in our dataset relates to every other. Darker red cells indicate stronger positive correlations and darker blue cells indicate negative correlations; the diagonal is hidden since a variable always perfectly correlates with itself. We expect physical inactivity to show the strongest positive correlation with obesity, and median income to show a negative correlation — higher-income counties tend to have lower obesity rates. If grocery store access is negatively correlated with fast food density, that would indicate the two variables partly capture the same underlying food environment, which we would account for in our regression model. Any predictor pair with r > 0.8 would signal multicollinearity concerns addressed in the Analysis section.
